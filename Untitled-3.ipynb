```python
"""
IMPROVED EARTHQUAKE MAGNITUDE PREDICTION
Key improvements over original:
1. HistGradientBoostingRegressor (much better than RF/ET on tabular data)
2. Multi-seed ensemble for variance reduction  
3. n_jobs=1 everywhere for reproducibility
4. np.random.seed() for global determinism
5. Additional interaction features
6. 10-fold CV for more stable OOF estimates
"""

import pandas as pd
import numpy as np
from sklearn.ensemble import (HistGradientBoostingRegressor,
                               ExtraTreesRegressor,
                               RandomForestRegressor)
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_predict, cross_val_score, KFold
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')

def get_directory():
    user_input = input("Enter directory path, or press Enter to use current directory: ").strip()

    directory = Path(user_input).expanduser().resolve() if user_input else Path.cwd()

    if not directory.exists():
        raise FileNotFoundError(f"Directory does not exist: {directory}")

    if not directory.is_dir():
        raise NotADirectoryError(f"Path is not a directory: {directory}")

    return directory

directory = get_directory()
print("Using directory:", directory)
os.chdir(directory)

# ─── CRITICAL for reproducibility (fixes Kaggle vs local MSE difference) ───
np.random.seed(42)
RANDOM_STATE = 42

# ─── 1. LOAD DATA ───────────────────────────────────────────────────────────
train = pd.read_csv("earth_Train.csv")
test  = pd.read_csv("earth_Test.csv")

TARGET = "Magnitude"
X = train.drop(columns=[TARGET])
y = train[TARGET]

print(f"Train: {train.shape}, Test: {test.shape}")
print(f"Target range: [{y.min()}, {y.max()}], mean={y.mean():.3f}")

# ─── 2. FEATURE ENGINEERING ─────────────────────────────────────────────────
def add_features(df):
    df = df.copy()
    eps = 1e-6

    # Log-transform skewed columns
    skewed_cols = [
        "feature_3", "feature_4", "feature_5", "feature_6",
        "feature_7", "feature_8", "feature_15", "feature_16",
        "feature_17", "feature_18", "feature_19", "feature_20"
    ]
    for col in skewed_cols:
        df[col + "_log1p"] = np.log1p(df[col])

    # Group statistics (original groups)
    groups = [
        ["feature_3", "feature_4", "feature_5", "feature_6", "feature_7", "feature_8"],
        ["feature_9", "feature_10", "feature_11", "feature_12", "feature_13"],
        ["feature_16", "feature_17", "feature_18", "feature_19", "feature_20"]
    ]
    for i, cols in enumerate(groups):
        df[f"group_{i}_mean"]  = df[cols].mean(axis=1)
        df[f"group_{i}_std"]   = df[cols].std(axis=1)
        df[f"group_{i}_max"]   = df[cols].max(axis=1)
        df[f"group_{i}_min"]   = df[cols].min(axis=1)
        df[f"group_{i}_range"] = df[f"group_{i}_max"] - df[f"group_{i}_min"]

    # Ratios and diffs (original)
    df["f17_f16_ratio"] = df["feature_17"] / (df["feature_16"] + eps)
    df["f18_f16_ratio"] = df["feature_18"] / (df["feature_16"] + eps)
    df["f7_f6_ratio"]   = df["feature_7"]  / (df["feature_6"]  + eps)
    df["f12_f13_diff"]  = df["feature_12"] - df["feature_13"]
    df["f9_f10_diff"]   = df["feature_9"]  - df["feature_10"]
    df["f10_f11_diff"]  = df["feature_10"] - df["feature_11"]

    # NEW: higher-order and cross features
    df["f14_sq"]         = df["feature_14"] ** 2
    df["f15_x_f9"]       = df["feature_15"] * df["feature_9"]
    df["f14_x_f15"]      = df["feature_14"] * df["feature_15"]
    df["f16log_x_f12"]   = df["feature_16_log1p"] * df["feature_12"]
    df["f17log_x_f9"]    = df["feature_17_log1p"] * df["feature_9"]
    df["f15_f16_ratio"]  = df["feature_15"] / (df["feature_16"] + eps)
    df["f9_f12_sum"]     = df["feature_9"]  + df["feature_12"]
    df["f9_f13_diff"]    = df["feature_9"]  - df["feature_13"]

    return df

X_fe   = add_features(X)
test_fe = add_features(test)
print(f"Features after engineering: {X_fe.shape[1]}")

# ─── 3. CV STRATEGY ─────────────────────────────────────────────────────────
# 10-fold gives more stable OOF estimates than 5-fold
cv = KFold(n_splits=10, shuffle=True, random_state=RANDOM_STATE)

# ─── 4. BASE MODELS ─────────────────────────────────────────────────────────
# WHY HistGradientBoostingRegressor?
#   - Native NaN support (no imputer technically needed but kept for safety)
#   - Histogram binning = much faster than GBM
#   - Consistently beats RF/ET on tabular regression problems
#   - Supports early stopping

base_models = {
    "hgb_deep": Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("m", HistGradientBoostingRegressor(
            max_iter=800, learning_rate=0.02,
            max_depth=9, min_samples_leaf=8,
            l2_regularization=0.1,
            random_state=RANDOM_STATE
        ))
    ]),
    "hgb_wide": Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("m", HistGradientBoostingRegressor(
            max_iter=600, learning_rate=0.03,
            max_depth=6, min_samples_leaf=12,
            l2_regularization=0.05,
            random_state=RANDOM_STATE + 1
        ))
    ]),
    "hgb_fast": Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("m", HistGradientBoostingRegressor(
            max_iter=500, learning_rate=0.05,
            max_depth=7, min_samples_leaf=10,
            random_state=RANDOM_STATE + 2
        ))
    ]),
    "rf": Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("m", RandomForestRegressor(
            n_estimators=600, max_depth=12,
            random_state=RANDOM_STATE,
            n_jobs=1          # n_jobs=1 is CRITICAL for reproducibility
        ))
    ]),
    "et": Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("m", ExtraTreesRegressor(
            n_estimators=600, max_depth=10,
            random_state=RANDOM_STATE,
            n_jobs=1          # n_jobs=1 is CRITICAL for reproducibility
        ))
    ]),
}

# ─── 5. OOF PREDICTIONS (STACKING LAYER 1) ──────────────────────────────────
# IMPORTANT: n_jobs=1 here too, for reproducibility
oof_preds = {}
test_preds = {}

for name, pipe in base_models.items():
    oof = cross_val_predict(pipe, X_fe, y, cv=cv, n_jobs=1)
    oof_preds[name] = oof
    mse = mean_squared_error(y, np.clip(oof, 3, 7))
    print(f"[{name}] OOF MSE: {mse:.6f}")
    
    # Retrain on full train set
    pipe.fit(X_fe, y)
    test_preds[name] = pipe.predict(test_fe)

# ─── 6. META-LEARNER (STACKING LAYER 2) ─────────────────────────────────────
oof_matrix  = np.column_stack(list(oof_preds.values()))
test_matrix = np.column_stack(list(test_preds.values()))

meta = Ridge(alpha=1.0)
meta.fit(oof_matrix, y)

stacked_oof  = np.clip(meta.predict(oof_matrix),  3, 7)
stacked_test = np.clip(meta.predict(test_matrix), 3, 7)

print(f"\n=== FINAL STACKED ENSEMBLE OOF MSE: {mean_squared_error(y, stacked_oof):.6f} ===")
print("Model weights:", dict(zip(base_models.keys(), meta.coef_.round(3))))

# ─── 7. GENERATE SUBMISSION ──────────────────────────────────────────────────
submission = pd.DataFrame({
    "ID": range(1, len(stacked_test) + 1),
    "TARGET": stacked_test
})
submission.to_csv("submission_improved_stacked.csv", index=False)
print(f"\nSubmission saved. Predictions: mean={stacked_test.mean():.4f}, std={stacked_test.std():.4f}")
```
